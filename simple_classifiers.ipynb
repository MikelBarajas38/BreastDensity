## KNN

```python
def knn(x, y, k):

    k_validation = 5

    k_model = Pipeline(
        steps=[("scaler", StandardScaler()), ('knn', KNeighborsClassifier(k, weights = 'distance'))]
        # steps=[('knn', KNeighborsClassifier(k, weights = 'distance'))]
    )

    scores = cross_val_score(k_model, x, y, cv=k_validation)

    accuracy = scores.mean()

    return accuracy, scores
```

```python
def get_best_k(x, y):
    model_pred = None
    best_scores = None
    best_accuracy = -1
    best_k = 0
    k_validation = 5

    for k in range(1, int(len(x)/k_validation)):

        k_accuracy, scores = knn(x, y, k)

        if best_accuracy < k_accuracy:
            best_accuracy = k_accuracy
            best_scores = scores
            best_k = k

    print()
    print(f'> best k: {best_k}')
    print(f'acc: {best_accuracy}')
    print(best_scores)
    print()

    return best_k, best_accuracy, best_scores
```

```python
def get_confusion_matrix(x, y, k):

    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)

    k_model = Pipeline(
        steps=[("scaler", StandardScaler()), ('knn', KNeighborsClassifier(k, weights = 'distance'))]
        # steps=[('knn', KNeighborsClassifier(k, weights = 'distance'))]
    )

    k_model.fit(x_train, y_train)
    y_pred = k_model.predict(x_test)

    return confusion_matrix(y_test, y_pred)
```

## All Features

### MLO

```python
X = glcm_mlo_df.drop(columns=['density'])
y = glcm_mlo_df['density']

k, accuracy, scores = get_best_k(X, y)
cm = get_confusion_matrix(X, y, k)
print(cm)
```

### CC

```python
X = glcm_cc_df.drop(columns=['density'])
y = glcm_cc_df['density']

k, accuracy, scores = get_best_k(X, y)
cm = get_confusion_matrix(X, y, k)
print(cm)
```

### Full dataset (MLO + CC)

```python
X = glcm_df.drop(columns=['density'])
y = glcm_df['density']

k, accuracy, scores = get_best_k(X, y)
cm = get_confusion_matrix(X, y, k)
print(cm)
```

## Feature-Optimization?

Testing if using a subset of the features gives better results. Basically brute-force, performing a complete search by testing 2^6 - 7 combinations (all combinations - combinations with 0-1 features)

```python
def get_feature_subset(X, bitmask):
    columns = X.columns
    subset = []

    for i in range(5):
        if bitmask & (1 << i):
            for j in range(4):
                #print(j, i, j * 5 + i)
                #print(columns[j * 5 + i])
                subset.append(columns[j * 5 + i])
    
    if bitmask & (1 << 5):
        #print(columns[20])
        subset.append(columns[20])

    return X[subset]
```

```python
X = glcm_df.drop(columns=['density'])
y = glcm_df['density']

best_accuracy = -1
best_mask = 0

results = []

for i in range(1, 2**6):

    # skip if there are less than 2 features
    if(bin(i).count('1') < 2):
        continue

    temp = {}
    
    print(f'bitmask: {i} - {bin(i)}')
    X_subset = get_feature_subset(X, i)
    # print(f'features: {X_subset.columns}')
    
    k, accuracy, scores = get_best_k(X_subset, y)
    #accuracy, scores = knn(X_subset, y, k = 115)

    temp['bitmask'] = i
    temp['features'] = X_subset.columns
    temp['k'] = k
    temp['accuracy'] = accuracy
    temp['scores'] = scores
    temp['timestamp'] = pd.Timestamp.now()

    results.append(temp)

    if best_accuracy < accuracy:
        best_accuracy = accuracy
        best_mask = i

knn_optimization = pd.DataFrame(results)
knn_optimization.to_csv('knn.csv', index=False)

print(f'best mask: {best_mask} - {bin(best_mask)}')
print(f'features: {get_feature_subset(X, best_mask).columns}')
print(f'best accuracy: {best_accuracy}')
```
